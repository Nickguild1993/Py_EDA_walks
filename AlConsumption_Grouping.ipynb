Alcohol Consumption Data Set Grouping Exercise

```python
# import libraries

import pandas as pd
import numpy as np
import random

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import the dataset from the raw url

df = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv")
df.head(5)
```

```python
print(df.shape)
print(df["continent"].nunique())
```

```python
# which continent drinks more beer on average
# need to groupy by and then find max

group = df.groupby("continent").agg({"beer_servings": ["mean", "max"],
                                     "spirit_servings" : ["mean", "max"],
                                     "wine_servings" :["mean", "max"]
                                     })

group = group.reset_index()

print(group)
print("Europe drinks the most beer on average")
```

Get the mean alcohol consumption of every column by continent 

```python
df.head(1)
```

```python
mean_df = df.groupby("continent").agg({"beer_servings": "mean",
                                       "spirit_servings": "mean",
                                       "wine_servings": "mean",
                                       "total_litres_of_pure_alcohol": "mean"})
mean_df
```

creating a new column for total servings and reordering df a bit

```python
df.head(1)
```

```python
total_alcohol = df["beer_servings"] + df["spirit_servings"]+ df["wine_servings"]
```

```python
total_df = df
total_df.insert(4, "total_servings", total_alcohol)
total_df.tail(5)
```

```python
# reorder the columns

total_df = total_df[["country", "continent", "beer_servings", "spirit_servings", "wine_servings", "total_servings", "total_litres_of_pure_alcohol"]]
total_df.head(5)
```

get the median alcohol servings by continent

```python
median_df = total_df.groupby("continent").agg({"total_servings" : "median"})
median_df
```

```python
median_df.sort_values(by="total_servings", ascending = False)
```
