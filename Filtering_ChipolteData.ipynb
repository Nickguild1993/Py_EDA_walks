Various filter and sort functions on chipolte ds

```python
# get libraries
import pandas as pd
import numpy as np
import random
```

```python
# get dataset | raw github url w/ wonky delim

df = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv", delimiter ="\t")
df.head()
```

```python
# get rid of teh "$" in item_price and turn to float

df["item_price"] = df["item_price"].str.slice(1).astype(float)
df.tail(5)
```

```python
# all orders > 10
over_10 = df[df["item_price"] > 10]
over_10[:5]
```

```python
# new df with only item_name and item_price
df_slice = df[["item_name", "item_price"]]
df_slice[:5]
```

```python
# sort by the name of each item

df.sort_values(by="item_name", ascending=False)
```

```python
# quantity of most expensive item ordered

df.sort_values(by="item_price", ascending=False)[:3]
```

```python
# how many times was the veggie salad bowl ordered

df[df["item_name"] == "Veggie Salad Bowl"].sum()
```

```python
# how many times was more than 1 canned soda ordered?

df[(df["item_name"] == "Canned Soda") & (df["quantity"] > 1)].count()
```
