Regiment Data set | exercise on pivot tables, cross tabulation, and group bys

```python
# import libraries

import pandas as pd
import numpy as np 
import random
```

```python
# creating dataset from a dictionary
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'], 
        'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'], 
        'name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze', 'Jacon', 'Ryaner', 'Sone', 'Sloan', 'Piger', 'Riani', 'Ali'], 
        'preTestScore': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
        'postTestScore': [25, 94, 57, 62, 70, 25, 94, 57, 62, 70, 62, 70]}

df = pd.DataFrame(raw_data)
df.head(5)
```

```python
# the usual explorations
print("rows by columns:", df.shape)
print("--------------------------------")
print(df.dtypes)
print("--------------------------------")
print(df.isnull().sum())
```

Getting mean preTestscore for nighthawks

```python
nighthawks = df.groupby(["regiment", "company"]).agg({"preTestScore": "mean",
                                         "postTestScore": "mean"})
nighthawks
```

Mean of the combined company's preTestScore

```python
print("The mean preTestScore for all company's is:", nighthawks["preTestScore"].mean())
```

```python
df.groupby("company").describe()
```

present mean preTestScores grouped by regiment and company without heirarchical indexing 

```python
nighthawks.reset_index()
```

Group the entire dataframe by regiment and company

```python
groupby = df.groupby(["company", "regiment"])
```

```python
groupby.size()
```

Iterate over a group and print the name and whole data from regiment

```python
for group_labels, group_dataframe in groupby:
  print(group_dataframe)
  break
```
