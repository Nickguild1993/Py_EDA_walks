Statistical analysis with baby name dataset

```python
# import libraries
import pandas as pd
import numpy as np
import random
from random import random
import statistics 
```

```python
# grab the github raw url
# comma delimiter

df = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv")
df.head(5)
```

```python
# get rid of first two bogus columns
babies = df[["Name", "Year", "Gender", "State", "Count"]]
babies.head(3)
```

```python
babies[babies["Name"] == "Emma"]
```

```python
babies.shape
```

```python
babies["Gender"].value_counts()
```

```python
del babies["Year"]

names = babies.groupby("Name").sum()

names[:5]
```

```python
names["Count"].sum()
```

```python
names.sort_values(by="Count", ascending= False)[:5]
```

```python
# number of unique names

len(names)
```

Name with most occurances

```python
names["Count"].idxmax()
```

```python
# or 

names[names["Count"] == names["Count"].max()]
```

How many names are the least occuring

```python
names[names["Count"] == names["Count"].min()]
```

```python
# what is the median name occurance?

names[names["Count"] == names["Count"].median()]
```

```python
# what is the standard deviation of names
names["Count"].std()
```

```python
names["Count"].describe()
```
