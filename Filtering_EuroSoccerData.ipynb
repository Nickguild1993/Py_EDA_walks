Filtering and sorting euro 12 soccer data

```python
# import libraries

import pandas as pd
import numpy as np
import random
```

```python
# get the data from github raw url

df = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv")
df.head(5)
```

```python
# how many countries bruv
df["Team"].count()
```

```python
# subset of df

discipline = df[["Team", "Yellow Cards", "Red Cards"]]
discipline.sample(5)
```

```python
# sort by red cards then yellows
df.sort_values(by=["Red Cards", "Yellow Cards"], ascending = False)
```

```python
# filter teams that scored than 6 goals
six_goals = df[df["Goals"] > 6]
six_goals
```

```python
# filter for 4 goals aaaaaaand offsides

df[(df["Goals"] > 4) & (df["Offsides"] < 10)]
```

```python
# filter for teams that start with g

df[df["Team"].str.startswith("G")]
```
