Stack, Unstack, Melt, Pivot exercise

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head(3)
```

What is the male to female ratio on different days of the week?

```python
tips_groupby = tips.groupby(["day", "sex"]).agg({"size" : "sum"})
tips_groupby
```

#### Unstack

Moves sub index into columns

```python
unstack_tips = tips_groupby.unstack()
unstack_tips
```

Can do the same with the days of the week (moving the index to a column)

```python
tips_groupby.unstack(0)
```

output is multindex columns :/ 
- to fix

```python
# transform them boys
unstack_tips_copy = unstack_tips.copy()

unstack_tips_copy.columns = ["__".join(col).strip() for col in unstack_tips_copy.columns.values]
```

```python
unstack_tips_copy
```

```python
unstack_tips_copy["size__Male"] / unstack_tips_copy["size__Female"]
```

#### Stack

does the opposite- multindex in the columns or a long list of columns and puts them into the index

```python
# here is what it looks like before
tips_groupby.unstack()
```

now with stack

```python
unstack_tips.stack()
```

```python
unstack_tips.stack(0)
```

#### Melting and Pivoting

```python
cheese = pd.DataFrame({"first": ["John", "Terry"],
                       "last": ["Smith", "Wallace"],
                       "height": [5.6, 6.0],
                       "weight" : [126,190]})
cheese
```

**Melt** does stacking in one operation

```python
cheese.melt(id_vars = ["first", "last"])
```

Can do the exact same thing with stacking though
except the columns names but obviously ya can change em

```python
cheese.set_index(["first", "last"], inplace = True)
cheese.stack().reset_index()
```
