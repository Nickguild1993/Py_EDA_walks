merging / joining housing market dataset

```python
# import libraries

import pandas as pd
import numpy as np
import random
from random import random
```

Create 3 seperate series (all have 100 obs)
- random number from 1 - 4
- random number from 1 - 3
- random number from 10,000 - 30,000

```python
s1 = np.random.randint(1,5, 100)

s2 = np.random.randint(1,4,100)

s3 = np.random.randint(100, 301, 100)
```

create a dataframe from the series

```python
s1.dtype
```

```python
df = pd.DataFrame({"beds": s1,
                   "baths": s2,
                   "price_sqr_meter":s3})
df
```

```python
# lets add a new column

s4 = np.random.randint(1000,4001, 100)

df.insert(3, "total_square_footage", s4)

df
```

```python
df = df.rename(columns = {"price_sqr_meter" : "price_square_foot"})
df.head(3)
```

```python
# add a computed column

df["price"] = df["price_square_foot"] * df["total_square_footage"]
df.head(3)
```

```python
df["price"] = df["price"].map("${0:,.2f}".format)
```
